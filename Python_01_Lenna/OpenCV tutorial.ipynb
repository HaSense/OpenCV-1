```python
import numpy as np
import cv2
import os
from matplotlib import pyplot as plt

imageFile = 'lenna.bmp'

# 파일이 있는지 확인하는 방법if os.path.isfile(imageFile):	# 파일이 있는지 확인하기
if os.path.isfile(imageFile):
    img = cv2.imread(imageFile) # numpy 배열이 넘어옴에 주의
else:
    print("Image load failed!")
    exit()

# matplotlib는 RGB임. C++ 방식에서는 BGR 방식 사용함.
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```
